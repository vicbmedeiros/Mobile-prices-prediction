```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')
```

```python
df_test.head()
```

```python
df_train.columns.tolist()
```

```python
X_train = df_train.drop('price_range', axis=1)
y_train = df_train['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_val)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_val.to_numpy().reshape(len(y_val),1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_val, y_pred)
accu_s = accuracy_score(y_val,y_pred)
print(cm) #65 right of who bought, and 24 right of doesnt bought
print(accu_s)
class_report = classification_report(y_val, y_pred)
print(class_report)
```
